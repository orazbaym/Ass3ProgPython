# Task 1

```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
streets = ['Абая', 'Саина', 'Кылышбаевна', 'Абульхаирхана', 'Мангилик']
statusValue = ['да', 'нет']
bathrooms = ['раздельный', 'совместный']
owners = ['Дуйсенбай', 'Сарсенбай', 'Жумабай', 'Сенбай', 'Бисембай']
```

```python
data = {
    "Улица": [random.choice(streets) for _ in range(200)],
    "Дом": [f"Дом {i}" for i in range(1, 201)],
    "Расстояние до центра (км)": [random.randint(1, 10) if random.random() > 0.2 else None for _ in range(200)],
    "Площадь (кв. м)": [random.uniform(40, 120) for _ in range(200)],
    "Количество комнат": [random.randint(1, 4) for _ in range(200)],
    "Метраж жилой площади (кв. м)": [random.uniform(30, 80) for _ in range(200)],
    "Этаж": [random.randint(1, 20) for _ in range(200)],
    "Наличие лифта": [random.choice(statusValue) for _ in range(200)],
    "Санузел": [random.choice(bathrooms) if random.random() > 0.1 else None for _ in range(200)],
    "Кухня студия": [random.choice(statusValue) for _ in range(200)],
    "Цена в месяц (тыс. тг)": [random.randint(50, 200) if random.random() > 0.1 else None for _ in range(200)],
    "Дата аренды": [(datetime.now() - timedelta(days=random.randint(1, 365))).strftime("%Y-%m-%d") for _ in range(200)],
    "Имя владельца": [random.choice(owners) for _ in range(200)],
    "Контактный номер": [f"8{random.randint(700, 999)}-{random.randint(100, 999)}-{random.randint(1000, 9999)}" for _ in range(200)]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('rental_data.csv', index=False)
df
```

# Task 3

```python
df_updated['Price_per_sqm'] = df_updated['Price'] / df_updated['Area']
df_updated
```

```python
import numpy as np
```

```python
random_floors = np.random.randint(df_updated['Floor'], high=df_updated['Floor'] + 5, size=len(df_updated))
df_updated['Total_Floors'] = np.maximum(random_floors, df_updated['Floor'])
df_updated

```

```python
# Создание категории этажности
def floor_category(row):
    if row['Floor'] == 1:
        return 'Первый'
    elif row['Floor'] == row['Total_Floors']:
        return 'Последний'
    else:
        return 'Другой'

# Добавление столбца с категорией этажности
df_updated['Floor_Category'] = df_updated.apply(floor_category, axis=1)

df_updated
```

```python
# Вывод всех квартир на 1 этаже с учетом информации об общем количестве этажей
first_floor = df_updated[(df_updated['Floor'] == 1) & (df_updated['Total_Floors'] >= 1)]
first_floor
```

```python
df_updated['Living_to_Total'] = df_updated['Living_area'] / df_updated['Area']
df_updated
```
